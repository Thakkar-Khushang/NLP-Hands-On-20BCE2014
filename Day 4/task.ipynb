```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(binary=True)
```

```python
vect.fit(corpus)
vocab = vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}: {}".format(key, vocab[key]))
```
